```python
import pandas as pd
import os
```

```python
def concat_file(dir_root, begin, end):
    data_all = []
    for year in range(begin, end):
        data_name = os.path.join(dir_root+str(year), 'data_bq_drop.csv')
        data = pd.read_csv(data_name, index_col=0)
        data = data.iloc[:, :-1]
        data_all.append(data)
    result = pd.concat(data_all, axis=1)
    return result
```

```python
# 将多个年份的数据垂直合并
def get_bach(data_root, begin, end):
    data_list = []
    for year in range(begin, end):
        data_name = os.path.join(data_root+str(year), 'data_bq_drop.csv')
        data = pd.read_csv(data_name, index_col=0)
        data_list.append(data)
    data_concat = pd.concat(data_list)
    return data_concat
```

```python
data = get_bach('/data/file/classification_data/2012-2019/data_sum/', 2012, 2018)
print(data)
```

```python
# 将带有标签的数据分成数据和标签两个文件
def divide(data_root, filename):
    data_name = os.path.join(data_root, filename)
    data = pd.read_csv(data_name, index_col=0)
    print(data)
    train_data = data.iloc[:, :-1]
    train_label = data.iloc[:, -1:]
    print(train_data)
    print(train_label)
    print()
    train_data.to_csv(os.path.join(data_root, 'train_'+filename))
    train_label.to_csv(os.path.join(data_root, 'label_'+filename))
```
